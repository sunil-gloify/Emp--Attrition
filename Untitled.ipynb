```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2
```

```python
def perform_chi_sqaure_test(var_1, var_2):
    contingency_table = pd.crosstab(train_data[var_1],train_data[var_2])
    observed_value = contingency_table.values
    b = chi2_contingency(contingency_table)
    excepted_values = b[3]

    #degree of freedom
    no_of_rows = len(contingency_table.iloc[0:,0])
    no_of_columns = len(contingency_table.iloc[0,0:])
    degree_f = (no_of_rows-1) * (no_of_columns-1)
    print("Degree Of Freedom : ",degree_f)

    #significant 5%
    alpha = 0.05
    print("Significant level :", alpha)

    #chi_square statistic
    chi_square = sum([(o-e)**2./e for o,e in zip(observed_value, excepted_values)])
    chi_square_statistic = chi_square[0] + chi_square[1]
    print("chi_square_statistic : ",chi_square_statistic)

    #crictical value
    crictical_value = chi2.ppf(q=1-alpha,df= degree_f)
    print("crictical_value : ", crictical_value)

    # p value
    p_value = 1-chi2.cdf(x=chi_square_statistic, df=degree_f)
    print("p_value", p_value)

    if chi_square_statistic>=crictical_value:
        print("Reject Ho, There is relationship between 2 categorical variable")
    else:
        print("Retain Ho, There is no relationship between 2 categorical variable")

    if p_value<=alpha:
        print("Reject Ho, There is relationship between 2 categorical variable")
    else:
        print("Retain Ho, There is no relationship between 2 categorical variable")
```

```python
for x in categorical_list:
    for i in categorical_list:
        if i != x:
            print('chi square test on : ', x, '', i, '\n')
            perform_chi_sqaure_test(x,i)
            print('----------------------------------------/n')
```

Anova Test

```python
def perform_anova_test(x, y):
    train_anova = train_data[[x, y]]
    groups = train_anova.groupby(x).count().reset_index()
    print(groups)

    unique_majors = train_anova[x].unique()
    for major in unique_majors:
        stats.probplot(train_anova[train_anova[x] == major][y], dist="norm", plot=plt)
        plt.title("Probability Plot - " + str(major))
        plt.show()

    ratio = train_anova.groupby(x).std().max() / train_anova.groupby(x).std().min()
    print(ratio)

    # Create ANOVA backbone table
    data = [['Between Groups', '', '', '', '', '', ''],
            ['Within Groups', '', '', '', '', '', ''],
            ['Total', '', '', '', '', '', '']]
    anova_table = pd.DataFrame(data, columns=['Source of Variation', 'SS', 'df', 'MS', 'F', 'P-value', 'F-crit'])
    anova_table.set_index('Source of Variation', inplace=True)

    # Calculate SSTR and update ANOVA table
    x_bar = train_anova[y].mean()
    SSTR = train_anova.groupby(x).count() * (train_anova.groupby(x).mean() - x_bar)**2
    anova_table['SS']['Between Groups'] = SSTR[y].sum()

    # Calculate SSE and update ANOVA table
    SSE = (train_anova.groupby(x).count() - 1) * train_anova.groupby(x).std()**2
    anova_table['SS']['Within Groups'] = SSE[y].sum()

    # Calculate SSTotal and update ANOVA table
    SSTotal = SSTR[y].sum() + SSE[y].sum()
    anova_table['SS']['Total'] = SSTotal

    # Update degrees of freedom
    anova_table['df']['Between Groups'] = train_anova[x].nunique() - 1
    anova_table['df']['Within Groups'] = train_anova.shape[0] - train_anova[x].nunique()
    anova_table['df']['Total'] = train_anova.shape[0] - 1

    # Calculate MS
    anova_table['MS'] = anova_table['SS'] / anova_table['df']

    # Calculate F
    F = anova_table['MS']['Between Groups'] / anova_table['MS']['Within Groups']
    anova_table['F']['Between Groups'] = F

    # Calculate p-value
    anova_table['P-value']['Between Groups'] = 1 - stats.f.cdf(F, anova_table['df']['Between Groups'], anova_table['df']['Within Groups'])

    # Calculate F critical value
    alpha = 0.5
    tail_hypothesis_type = 'two-tailed'
    if tail_hypothesis_type == 'two-tailed':
        alpha /= 2
    anova_table['F-crit']['Between Groups'] = stats.f.ppf(1 - alpha, anova_table['df']['Between Groups'], anova_table['df']['Within Groups'])

    print(anova_table)

    # The p-value approach
    print("Approach 1: The p-value approach to hypothesis testing in the decision rule")
    conclusion = "Failed to reject the null hypothesis"
    if anova_table['P-value']['Between Groups'] <= alpha:
        conclusion = 'Null hypothesis is rejected'
    print('F-Score is:', anova_table['F']['Between Groups'], "P-value score is:", anova_table['P-value']['Between Groups'])
    print(conclusion)

    # The critical value approach
    print("Approach 2: The critical value approach to hypothesis testing in the decision rule")
    conclusion = "Failed to reject the null hypothesis"
    if anova_table['F']['Between Groups'] > anova_table['F-crit']['Between Groups']:
        conclusion = 'Null hypothesis is rejected'
    print('F-Score is:', anova_table['F']['Between Groups'], "Critical score is:", anova_table['F-crit']['Between Groups'])
    print(conclusion)
```

```python
class OneHotEncoder(SklearnOneHotEncoder):
    def __init__(self, **kwargs):
        super(OneHotEncoder, self).__init__(**kwargs)
        self.fit_flag = False

    def fit(self, X, **kwargs):
        out = super().fit(X)
        self.fit_flag = True
        return out

    def transform(self, X, **kwargs):
        sparse_matrix = super().transform(X)
        new_columns = self.get_new_columns(X=X)
        d_out = pd.DataFrame(sparse_matrix.toarray(), columns=new_columns, index=X.index)
        return d_out

    def fit_transform(self, X, **kwargs):
        self.fit(X)
        return self.transform(X)

    def get_new_columns(self, X):
        new_columns = []
        for i, column in enumerate(X.columns):
            for category in self.categories_[i]:
                new_columns.append(f'{column}_{category}')
        return new_columns
        
```

```python
import math
from sklearn.metrics import median_absolute_error, mean_absolute_error
from sklearn import metrics

def rmse(x, y):
    return math.sqrt(((x - y) ** 2).mean())

def print_score(m):
    res = [rmse(m.predict(X_train), Y_train),
           rmse(m.predict(X_test), Y_test),
           m.score(X_train, Y_train),
           m.score(X_test, Y_test),
           median_absolute_error(m.predict(X_train), Y_train),
           median_absolute_error(m.predict(X_test), Y_test),
           mean_absolute_error(m.predict(X_train), Y_train),
           mean_absolute_error(m.predict(X_test), Y_test)
          ]

    print("RMSE-Train: " + str(res[0]) +
          "\nRMSE-Test: " + str(res[1]) +
          "\nScore-Train: " + str(res[2]) +
          "\nScore-Test: " + str(res[3]) +
          "\nMedAE-Train: " + str(res[4]) +
          "\nMedAE-Test: " + str(res[5]) +
          "\nMean-AE-Train: " + str(res[6]) +
          "\nMean-AE-Test: " + str(res[7]))
```

```python

def visualize_importance(feature_importance, feat_train_df):
    feature_importance_df = pd.DataFrame()
    _df = pd.DataFrame()
    _df['feature_importance'] = feature_importance
    _df['column'] = feat_train_df.columns
    feature_importance_df = pd.concat([feature_importance_df, _df], axis=0, ignore_index=True)

    order = feature_importance_df.groupby('column') \
        .sum()[['feature_importance']] \
        .sort_values('feature_importance', ascending=False).index[:50]

    fig, ax = plt.subplots(figsize=(8, max(6, len(order) * 0.25)))
    sns.boxenplot(data=feature_importance_df,
                  x='feature_importance',
                  y='column',
                  order=order,
                  ax=ax,
                  palette='viridis',
                  orient='h'
                  )
    ax.tick_params(axis='x', rotation=0)
    ax.set_title('Importance')
    ax.grid()
    fig.tight_layout()
    return fig, ax
    
```

```python
def compare_models(models, names, x_train, y_train, x_test, y_test):
    data = {'Metric': ['RMSE', 'MedAE', 'MAE', 'R-squared']}
    df_train = pd.DataFrame(data)
    df_test = pd.DataFrame(data)
    
    def rmse(x, y):
        return math.sqrt(((x - y) ** 2).mean())
    
    for model, name in zip(models, names):
        y_pred_train = model.predict(x_train)
        y_pred_test = model.predict(x_test)
        
        res = [
            rmse(y_pred_train, y_train),
            rmse(y_pred_test, y_test),
            metrics.median_absolute_error(y_pred_train, y_train),
            metrics.median_absolute_error(y_pred_test, y_test),
            metrics.mean_absolute_error(y_pred_train, y_train),
            metrics.mean_absolute_error(y_pred_test, y_test),
            metrics.r2_score(y_pred_train, y_train),
            metrics.r2_score(y_pred_test, y_test)
        ]
        
        df_train[name] = [res[0], res[2], res[4], res[6]]
        df_test[name] = [res[1], res[3], res[5], res[7]]
        
    return df_train, df_test
```
